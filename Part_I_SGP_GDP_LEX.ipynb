```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pandas_datareader import wb
from datetime import datetime
```

```python
gdp_all_1960_raw = wb.download(indicator=["NY.GDP.PCAP.CD"], country="all", start=1960, end=1960)

# selecting countries
gdp_all_1960_country = gdp_all_1960_raw.iloc[47:] 
gdp_all_1960 = gdp_all_1960_country.dropna()

# renaming column
gdp_all_1960.columns = ["GDP_per_capita"]
print(gdp_all_1960.head())
print(gdp_all_1960.xs("Singapore", level=0))
```

```python
gdp_all_1960_quartiles = np.quantile(gdp_all_1960["GDP_per_capita"], [0.25, 0.5, 0.75])
mean_gdp_all_1960 = np.mean(gdp_all_1960["GDP_per_capita"])
median_gdp_all_1960 = np.median(gdp_all_1960["GDP_per_capita"])

print("GDP quartiles for all countries in 1960 in US$: ${:.2f}, ${:.2f} and ${:.2f}.".\
      format(gdp_all_1960_quartiles[0], gdp_all_1960_quartiles[1], gdp_all_1960_quartiles[2]))
print("Mean GDP for all countries in 1960 : ${:.2f}.".format(mean_gdp_all_1960))
print("Median GDP for all countries in 1960 : ${:.2f}.".format(median_gdp_all_1960))

gdp_all_1960_percent_67 = np.percentile(gdp_all_1960, 67)

print("SGP's GDP of $428.06 in 1960 lies in 3rd quartile of all countries - \
at 67th(${:.2f}) percentile.".format(gdp_all_1960_percent_67))
```

```python
x = range(1960, 2019)
gdp_sgp = gdp_lex_sgp["GDP_per_capita"]
lex_sgp = gdp_lex_sgp["Life_expectancy"]

plt.figure(figsize = (8,6))

ax1 = plt.subplot()
ax1.plot(x, gdp_sgp, color = "tab:orange", linewidth=2, label="GDP per capita")
ax1.set_ylabel("GDP per capita")
ax1.spines["left"].set_color("tab:orange")
ax1.yaxis.label.set_color('tab:orange')
ax1.tick_params(axis='y', colors='tab:orange')
ax1.annotate("--- GDP per capita", c="tab:orange", xy=(1960, 60000))
ax1.annotate("Gradual gradient", c="tab:orange", xy=(1982, 5000), xytext=(1999, 0), \
             arrowprops=dict(facecolor='black', width=1, shrink=0.06))
ax1.annotate("Accelerated \ngradient", c="tab:orange", xy=(2008, 35000), xytext=(2010, 20000), \
             arrowprops=dict(facecolor='black', width=1, shrink=0.06))

ax2 = ax1.twinx()
ax2.plot(x, lex_sgp, color = "tab:green", linewidth=2, label="Life Expectancy" )
ax2.set_ylabel("Life Expectancy")
ax2.spines["right"].set_color("tab:green")
ax2.yaxis.label.set_color('tab:green')
ax2.tick_params(axis='y', colors='tab:green')
ax1.annotate("--- Life Expectancy", c="tab:green", xy=(1960, 55000))

ax1.set_xlabel("Year")
plt.title("GDP and Life Expectancy - SGP", fontweight="bold")
plt.show()

print("Coefficient Matrix: {}".format(np.corrcoef([gdp_lex_sgp["Life_expectancy"], gdp_lex_sgp["GDP_per_capita"]])))
```
